```python
def find_local_extrema(data):
    peaks = []
    valleys = []
    # Handling the first element
    if data[0] > data[1]:
        peaks.append(0)
    if data[0] < data[1]:
        valleys.append(0)
    # Handling elements in between
    for i in range(1, len(data) - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            peaks.append(data[i])
        if data[i] < data[i - 1] and data[i] < data[i + 1]:
            valleys.append(data[i])
    # Handling the last element
    if data[-1] > data[-2]:
        peaks.append(data[len(data)-1])
    if data[-1] < data[-2]:
        valleys.append(data[len(data) - 1])
    return [*peaks, *valleys]
```

```python
import pandas as pd

df = pd.read_csv("3months.csv")
df.head()
```

```python
x = df.close.values
print(len(x))
for i in range(12):
    x = find_local_extrema(x)
    print(len(x))
```

```python
# Modifying the find_local_extrema function to return only the extrema values
def find_local_extrema(data):
    extrema = []
    # Handling elements in between
    for i in range(1, len(data) - 1):
        if (data[i] > data[i - 1] and data[i] > data[i + 1]) or (data[i] < data[i - 1] and data[i] < data[i + 1]):
            extrema.append(data[i])
    # Handling the first element
    if data[0] > data[1] or data[0] < data[1]:
        extrema.append(data[0])
    # Handling the last element
    if data[-1] > data[-2] or data[-1] < data[-2]:
        extrema.append(data[-1])
    return extrema

# Applying the function iteratively until only 4 extrema are left
x = df['close'].values
extrema_counts = [len(x)]
for i in range(4):
    print(len(x))
    x = find_local_extrema(x)
    extrema_counts.append(len(x))

extrema_counts
```
