```python
import os
import pandas as pd
extract_path = r"C:\Users\xyz\Desktop\bodi2"

ohlcv_data_new = pd.read_csv(os.path.join(extract_path, "1h_ETH_withFeatures_v2.csv"))
ohlcv_data_new
```

```python

# Function to calculate the True Range
def true_range(high, low, prev_close):
    return max(high - low, abs(high - prev_close), abs(low - prev_close))

# Calculating the True Range for each row
ohlcv_data_new['true_range'] = ohlcv_data_new.apply(lambda row: true_range(row['High'], row['Low'], ohlcv_data_new.loc[row.name - 1, 'Close'] if row.name > 0 else 0), axis=1)

# Calculating the Average True Range (ATR) using a rolling window of 14 periods
atr_period = 14
ohlcv_data_new['atr'] = ohlcv_data_new['true_range'].rolling(window=atr_period).mean()

# Displaying the first few rows with the true range and ATR
ohlcv_data_new.head(20)
```

```python
# Function to label a data point based on the given condition

atr_plus = 2
atr_minus = 1

def label_data(row):
    current_close = row['Close']
    current_atr = row['atr']
    future_closes = ohlcv_data_new.loc[row.name + 1:, 'Close'].values

    # Check for the condition: Close price rises by 2 ATR before falling by 1 ATR
    has_risen_2_atr = False
    for future_close in future_closes:
        price_diff = future_close - current_close
        if price_diff <= -atr_minus*current_atr:
            return False
        if price_diff >= atr_plus * current_atr:
            return True

# Apply the labeling function to each row
ohlcv_data_new['label'] = ohlcv_data_new.apply(label_data, axis=1)

# Displaying the first few rows with the label
```
